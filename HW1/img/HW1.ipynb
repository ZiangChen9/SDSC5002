Show first 8 rows of the dataset.

```python
import pandas as pd
df = pd.read_csv('./data/Housing.csv')
df.head(8)
```

Given the median and mean of crim.

```python
average_value = df['crim'].mean()
print(average_value)
```

Calculate 1st and 3rd quantile of medv.

```python
q1 = df['crim'].quantile(0.25)
q3 = df['crim'].quantile(0.75)
print(q1,q3)
```

Plot histogram of medv.

```python
import matplotlib.pyplot as plt
plt.title('histogram of medv')
plt.xlabel('median home price in different neighborhoods')
plt.hist(df['medv'],bins=20,edgecolor='black',alpha=0.7,)
```

On the same plot, draw histograms of medv for high crime rate (crim > 0.5)
and low crime rate (crim £ 0.5), respectively.

```python
high_crim_data = df[df['crim'] > 0.5]
low_crim_data = df[df['crim'] <= 0.5]
plt.figure(figsize=(14,6))
title = 'histograms of medv for different crime rate'
plt.suptitle(title, fontsize=16)
plt.subplot(1, 2, 1)
plt.hist(low_crim_data['medv'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Median Home Price in Different Neighborhoods')
plt.title('Low Crime Rate (crim <= 0.5)')
plt.subplot(1, 2, 2)
plt.hist(high_crim_data['medv'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Median Home Price in Different Neighborhoods')
plt.title('High Crime Rate (crim > 0.5)')
plt.tight_layout(rect=[0, 0, 1, 0.95])  # 调整布局，确保标题不会被裁剪
plt.show()
```
